````python
# Step's to handel null values

# step: 1
# Drope age_certification(93% null) & seasons(53% null) (not useful for analysis)
Amazon_Prime.drop(['age_certification','seasons'], axis= 1,inplace=True)

# step: 2
"""Drop null values for columns like [description,tmdb_popularity,person_id, name, role]
   bacause these all of individual null values are less then 1% ,
   Since impact is minimal, removing them avoids unnecessary imputation bias"""

Amazon_Prime.dropna(subset=['description','tmdb_popularity','person_id', 'name', 'role'], inplace = True)

# step: 3
# Replace imdb_id  null values with "unknown" as no logical replacement exists
Amazon_Prime['imdb_id']= Amazon_Prime['imdb_id'].fillna('unknown')

# step: 4
"""fill null values for imdb_score , imdb_votes , tmdb_score with there individual median,
   medain is most ideal approch based on (grouped by type) to preserve distribution and reduce skew impact
"""
Amazon_Prime[['imdb_score','imdb_votes','tmdb_score']] = (
    Amazon_Prime.groupby('type')[['imdb_score','imdb_votes','tmdb_score']]
    .transform(lambda x: x.fillna(x.median()))
)

# step: 5
# handel null values for character with context-aware imputation
"""Fill missing 'character' values using the most frequent role per person (grouped by person_id and name)
This preserves contextual relationships and avoids arbitrary imputation like 'Unknown'
Falls back to 'Unknown' only when no valid character information is available"""

Amazon_Prime['character'] = (
    Amazon_Prime.groupby(['person_id','name'])['character']
    .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
)
#---------------------
# Handel dtype
#---------------------
# Convert imdb_votes --- float64--->---- int64
Amazon_Prime['imdb_votes'] = Amazon_Prime['imdb_votes'].astype('int64')
# Convert person_id flot64---?--- text formate (string-like)
Amazon_Prime['person_id'] = Amazon_Prime['person_id'].astype('object')

# final dataset info
print(f"{Amazon_Prime.info()}")
````

````python
"""Test data cleaning most important for
Prevents duplicate categoriey and
Required for Machine Learning, models treat text as features:
"""
# Text cleaning with str()---lowercase()---strip()
# Step: 1 for title clean
import pandas as pd

def text_clean(text):
    if pd.isna(text):
        return ""

    text = str(text).lower().strip()

    # remove special characters
    for char in ['~','!','@','#','$','%','^','&','*','(',')','_','+','=','-','[',']','{','}','|','\\',':',';','"','<','>',',','.','/','?']:
        text = text.replace(char, '')


    return text
# Final Implement
for col in ['title','description','name',	'character']:
    Amazon_Prime[col] = Amazon_Prime[col].apply(text_clean)
#----------------------------------------------------------------
# Final adjusment
#----------------------------------------------------------------
# First for title
Amazon_Prime[['title','name','character']] =Amazon_Prime[['title','name','character']].apply(lambda x:x.str.title())

# description
Amazon_Prime['description']= Amazon_Prime['description'].apply(lambda x: x.capitalize())
````

````python
# Handel genres &	production_countries Final one
def clean_list_column(x):
    if pd.isna(x):
        return []

    x = str(x)

    # Remove brackets and quotes
    x = x.replace('[','').replace(']','').replace("'", "")

    # Split into list
    items = x.split(',')

    cleaned = []
    for item in items:
        item = item.lower().strip()

        # remove digits
        item = ''.join([i for i in item if not i.isdigit()])

        # convert to Title Case
        item = item.title()

        if item:
            cleaned.append(item)

    return cleaned

# Final Implementation
Amazon_Prime['genres'] = Amazon_Prime['genres'].apply(clean_list_column)
Amazon_Prime['production_countries'] = Amazon_Prime['production_countries'].apply(clean_list_column)
````

````python
#--------------------------------------------------------
# Handel duplicates in production_countries
#--------------------------------------------------------
mapping = {
    'Us': 'USA',
    'US': 'USA',
    'United States Of America': 'USA',

    'Gb': 'UK',
    'United Kingdom': 'UK',

    'In': 'India'
}

Amazon_Prime['production_countries'] = Amazon_Prime['production_countries'].apply(
    lambda lst: [mapping.get(i, i) for i in lst]
)

Amazon_Prime['production_countries'] = Amazon_Prime['production_countries'].apply(
    lambda lst: list(set([mapping.get(i, i) for i in lst]))
)
````

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# First: Create AMprime_Score
# Represents a combined rating (0–10 scale) using IMDb and TMDb scores
Amazon_Prime['AMprime_Score'] = round(
    Amazon_Prime[['imdb_score','tmdb_score']].mean(axis=1), 2
)


# Second: Create AMprime_Popularity
"""
To represent popularity using IMDb votes and TMDb popularity:
- Applied log transformation to imdb_votes to reduce skewness
- Normalized both features to a 0–10 scale using Min-Max scaling
- Combined them using a balanced average
"""

# Step 1: Log transform IMDb votes
Amazon_Prime['votes_log'] = np.log1p(Amazon_Prime['imdb_votes'])

# Step 2: Normalize TMDb popularity
Amazon_Prime['tmdb_pop_norm'] = (
    (Amazon_Prime['tmdb_popularity'] - Amazon_Prime['tmdb_popularity'].min()) /
    (Amazon_Prime['tmdb_popularity'].max() - Amazon_Prime['tmdb_popularity'].min())
) * 10

# Step 3: Normalize votes
Amazon_Prime['votes_norm'] = (
    (Amazon_Prime['votes_log'] - Amazon_Prime['votes_log'].min()) /
    (Amazon_Prime['votes_log'].max() - Amazon_Prime['votes_log'].min())
) * 10

# Step 4: Final popularity score
Amazon_Prime['AMprime_Popularity'] = round(
    0.5 * Amazon_Prime['tmdb_pop_norm'] +
    0.5 * Amazon_Prime['votes_norm'], 2
)

# Step 5: Cleanup
Amazon_Prime.drop(['votes_log', 'tmdb_pop_norm', 'votes_norm'], axis=1, inplace=True)


# Third: Create final AMprime_Rating
"""
Final score combining:
- 60% weight to quality (AMprime_Score)
- 40% weight to popularity (AMprime_Popularity)
"""
Amazon_Prime['AMprime_Rating'] = round(
    0.6 * Amazon_Prime['AMprime_Score'] +
    0.4 * Amazon_Prime['AMprime_Popularity'], 2
)


# Fourth: Categorize release_year into eras
def categorize_era(year):
    if year < 1980:
        return 'Classic Era'
    elif year < 2000:
        return 'Old Era'
    elif year < 2010:
        return 'Early 2000s'
    elif year < 2020:
        return 'Modern Era'
    else:
        return 'Recent Era'

Amazon_Prime['Era'] = Amazon_Prime['release_year'].apply(categorize_era)

# Content Quality Category based on (AMprime_Rating)
Amazon_Prime['Content_Quality'] = 'Average'

# Movies
Amazon_Prime.loc[(Amazon_Prime['type'] == 'MOVIE') & (Amazon_Prime['AMprime_Rating'] >= 7.5), 'Content_Quality'] = 'Excellent'
Amazon_Prime.loc[(Amazon_Prime['type'] == 'MOVIE') & (Amazon_Prime['AMprime_Rating'] >= 6.5) & (Amazon_Prime['AMprime_Rating'] < 7.5), 'Content_Quality'] = 'Good'
Amazon_Prime.loc[(Amazon_Prime['type'] == 'MOVIE') & (Amazon_Prime['AMprime_Rating'] >= 5) & (Amazon_Prime['AMprime_Rating'] < 6.5), 'Content_Quality'] = 'Average'
Amazon_Prime.loc[(Amazon_Prime['type'] == 'MOVIE') & (Amazon_Prime['AMprime_Rating'] < 5), 'Content_Quality'] = 'Poor'

# Shows
Amazon_Prime.loc[(Amazon_Prime['type'] == 'SHOW') & (Amazon_Prime['AMprime_Rating'] >= 8), 'Content_Quality'] = 'Excellent'
Amazon_Prime.loc[(Amazon_Prime['type'] == 'SHOW') & (Amazon_Prime['AMprime_Rating'] >= 7) & (Amazon_Prime['AMprime_Rating'] < 8), 'Content_Quality'] = 'Good'
Amazon_Prime.loc[(Amazon_Prime['type'] == 'SHOW') & (Amazon_Prime['AMprime_Rating'] >= 5.5) & (Amazon_Prime['AMprime_Rating'] < 7), 'Content_Quality'] = 'Average'
Amazon_Prime.loc[(Amazon_Prime['type'] == 'SHOW') & (Amazon_Prime['AMprime_Rating'] < 5.5), 'Content_Quality'] = 'Poor'

# Categorized runtime into segments acc to type
def categorize_runtime(row):
    if row['type'] == 'MOVIE':
        if row['runtime'] < 90:
            return 'Short Movie'
        elif row['runtime'] <= 150:
            return 'Standard Movie'
        else:
            return 'Long Movie'
    else:  # SHOW
        if row['runtime'] < 30:
            return 'Short Episode'
        elif row['runtime'] <= 60:
            return 'Standard Episode'
        else:
            return 'Long Episode'

Amazon_Prime['runtime_category'] = Amazon_Prime.apply(categorize_runtime, axis=1)
````

````python
# Outliers deduction pipeline
def Outliers_deduction_func(df, group_col, target_col):

    IQR_df = df.groupby(group_col)[target_col].agg(
        Q1=lambda x: x.quantile(0.25),
        Q3=lambda x: x.quantile(0.75),
        Min='min',
        Max='max'
    )

    # IQR
    IQR_df['IQR'] = IQR_df['Q3'] - IQR_df['Q1']

    # Bounds
    IQR_df['LowerBound'] = IQR_df['Q1'] - 1.5 * IQR_df['IQR']
    IQR_df['UpperBound'] = IQR_df['Q3'] + 1.5 * IQR_df['IQR']

    # Detect direction
    IQR_df['LeftOutlier'] = IQR_df['Min'] < IQR_df['LowerBound']
    IQR_df['RightOutlier'] = IQR_df['Max'] > IQR_df['UpperBound']

    # Final status
    def get_status(row):
        if row['LeftOutlier'] and row['RightOutlier']:
            return 'Both'
        elif row['LeftOutlier']:
            return 'Left'
        elif row['RightOutlier']:
            return 'Right'
        else:
            return 'No'

    IQR_df['status'] = IQR_df.apply(get_status, axis=1)

    # Convert True/False to Yes/No (optional)
    IQR_df['LeftOutlier'] = IQR_df['LeftOutlier'].map({True: 'Yes', False: 'No'})
    IQR_df['RightOutlier'] = IQR_df['RightOutlier'].map({True: 'Yes', False: 'No'})

    # Reorder columns
    IQR_df = IQR_df[
        ['Q1','Q3','IQR','Min','LowerBound','Max','UpperBound',
         'LeftOutlier','RightOutlier','status']
    ]

    return IQR_df

 # Outliers deduction For loop for numeric column
cols = ['runtime','imdb_score','tmdb_score','imdb_votes','tmdb_popularity']
for col in cols:
  Outliers_df = Outliers_deduction_func(Amazon_Prime,'type',col)

  print("=" * 60)
  print(f"🔹 Outlier Summary for: {col.upper()}")
  print("=" * 60)
  print(Outliers_df)
  print("\n")
````

````python
# Outliers Flaging

def overall_outlier_flag(df, group_col, cols):

    df = df.copy()
    outlier_flag = pd.Series(False, index=df.index)

    for col in cols:

        # Step 1: Calculate bounds per group
        bounds = df.groupby(group_col)[col].agg(
            Q1=lambda x: x.quantile(0.25),
            Q3=lambda x: x.quantile(0.75)
        )

        bounds['IQR'] = bounds['Q3'] - bounds['Q1']
        bounds['LowerBound'] = bounds['Q1'] - 1.5 * bounds['IQR']
        bounds['UpperBound'] = bounds['Q3'] + 1.5 * bounds['IQR']

        # Step 2: Map bounds to rows
        temp = df[[group_col, col]].merge(
            bounds[['LowerBound','UpperBound']],
            left_on=group_col,
            right_index=True,
            how='left'
        )

        # Step 3: Check outliers
        col_flag = (temp[col] < temp['LowerBound']) | (temp[col] > temp['UpperBound'])

        # Step 4: Combine flags (OR logic)
        outlier_flag = outlier_flag | col_flag

    # Final column
    df['Any_Outlier'] = outlier_flag.map({True:'Yes', False:'No'})

    return df

# Final implemantation
cols = ['runtime','imdb_votes','tmdb_popularity']

df_final = overall_outlier_flag(Amazon_Prime, 'type', cols)

# Implementation
Amazon_Prime['Any_Outlier'] = overall_outlier_flag(
    Amazon_Prime, 'type', cols
)['Any_Outlier']

# Filter non-outliers
df_no_outliers = Amazon_Prime[Amazon_Prime['Any_Outlier'] == 'No']
# Check Shape
total = Amazon_Prime.shape[0]
clean = df_no_outliers.shape[0]
outliers = total - clean

print("Original Shape:", Amazon_Prime.shape)
print("Without Outliers Shape:", df_no_outliers.shape)

print("Non-Outlier %:", round((clean / total) * 100, 2))
print("Outlier %:", round((outliers / total) * 100, 2))

# Normalized_AMprime data
Normalized_AMprime = Amazon_Prime[Amazon_Prime['Any_Outlier'] == 'Yes']
````

**Feature Engineering**

- **AMprime_Score:** Created by combining IMDb and TMDb ratings to represent overall content quality on a 0–10 scale.
- **AMprime_Popularity:** Derived using **log transformation** on IMDb votes and **Min-Max normalization** of both votes and TMDb popularity to capture user engagement.
- **AMprime_Rating:** Final composite score calculated using a **weighted average (60% quality, 40% popularity)** for a balanced ranking system.
- **Era & Content_Quality:** Release years were grouped into eras, and content was categorized into quality levels (Excellent, Good, Average, Poor) with different thresholds for movies and shows.
- **runtime_category:** split runtime into category according to
 1) **Movies:** *Sort, standard & long movies.*  
 2) **TV show:** *Sort, standard & long Episode*
- **Any_Outliers:** Flaging outliers data  related to every numeric variable.
####**Conclusion**

These engineered features enable deeper insights into content performance across time, type, and quality segments. They provide a strong foundation for creating impactful visualizations and identifying trends, patterns, and top-performing content in further analysis.

````python
# Chart - 6 visualization
#---------------------------------------------------------------------------------------------
# Multiploat matrix AMprimE features with (Era, Content Quality, Runtime Category, Popularity)
#---------------------------------------------------------------------------------------------

def plot_multi_analysis(df, value_col, title_name):

    df = df.copy()

    # Aggregations
    era_df = df.groupby(['Era','type'])[value_col].mean().reset_index()
    runtime_df = df.groupby(['runtime_category','type'])[value_col].mean().reset_index()

    colors = {'MOVIE': '#00A8E1', 'SHOW': '#FF9900'}

    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=(
            f"Bar: {title_name} vs Era",
            f"Box: {title_name} vs Content Quality",
            f"Line: {title_name} vs Runtime Category",
            f"Scatter: {title_name} vs Popularity"
        )
    )

    # Loop once for all plots
    for t in ['MOVIE', 'SHOW']:

        d_main = df[df['type'] == t]
        d_era = era_df[era_df['type'] == t]
        d_run = runtime_df[runtime_df['type'] == t]

        # Bar
        fig.add_bar(x=d_era['Era'], y=d_era[value_col],
                    name=t, marker_color=colors[t],
                    legendgroup=t, showlegend=True,
                    row=1, col=1)

        # Box
        fig.add_box(x=d_main['Content_Quality'], y=d_main[value_col],
                    name=t, marker_color=colors[t],
                    legendgroup=t, showlegend=False,
                    row=1, col=2)

        # Line
        fig.add_scatter(x=d_run['runtime_category'], y=d_run[value_col],
                        mode='lines+markers',
                        name=t,
                        line=dict(color=colors[t], width=3),
                        legendgroup=t, showlegend=False,
                        row=2, col=1)

        # Scatter
        fig.add_scatter(x=d_main['AMprime_Popularity'], y=d_main[value_col],
                        mode='markers',
                        name=t,
                        marker=dict(color=colors[t], size=6, opacity=0.6),
                        legendgroup=t, showlegend=False,
                        row=2, col=2)

    # Layout
    fig.update_layout(
        template='plotly_dark',
        plot_bgcolor='#0F171E',
        paper_bgcolor='#0F171E',
        font=dict(color='#E6F1F5'),
        height=700,
        width=1000,
        title=f"{title_name} Analysis Across Categories",
        title_x=0.5,
        legend=dict(orientation="h", y=0.5, x=0.5, xanchor="center")
    )

    fig.show()
````

````python
# Chart -8 visualization code
#----------------------------------------------------
# Reusable Function (dynamic metric)
#----------------------------------------------------

def prepare(df, col, value_col):

    df = df.explode(col)
    df = df[df[col] != ""]

    # reduce clutter
    top = df[col].value_counts().head(10).index
    df = df[df[col].isin(top)]

    agg = (
        df.groupby([col, 'type'])[value_col]
        .agg(
            Mean='mean',
            Q1=lambda x: x.quantile(0.25),
            Q3=lambda x: x.quantile(0.75)
        )
        .reset_index()
    )

    agg['Upper'] = agg['Q3'] + 1.5 * (agg['Q3'] - agg['Q1'])

    agg = agg.melt(
        id_vars=[col, 'type'],
        value_vars=['Mean', 'Q3', 'Upper'],
        var_name='Metric',
        value_name='Score'
    )

    agg['Label'] = agg['type'] + " | " + agg['Metric']

    order = (
        df.groupby(col)[value_col]
        .mean()
        .sort_values(ascending=False)
        .index
    )

    return agg, order


# ----------------------------------------------------
# Plot Function
# ----------------------------------------------------
def plot_metric(value_col, title_name):

    country_df, country_order = prepare(
        Amazon_Prime, 'production_countries', value_col
    )

    genre_df, genre_order = prepare(
        Amazon_Prime, 'genres', value_col
    )

    # Strong contrast colors
    colors = {
    # MOVIES → Blue shades
    'MOVIE | Mean':  '#37E8C1',
    'MOVIE | Q3':    '#B03CE8',
    'MOVIE | Upper': '#3588E4',

    # SHOWS → Orange shades
    'SHOW | Mean':   '#DE9632',
    'SHOW | Q3':     '#0C9E26',
    'SHOW | Upper':  '#EEB182'
}

    # ------------------------------------------------
    # Subplots
    # ------------------------------------------------
    fig = make_subplots(
        rows=1, cols=2,
        subplot_titles=("Production Countries", "Genres"),
        horizontal_spacing=0.12
    )

    # ------------------------------------------------
    # Add Bars
    # ------------------------------------------------
    for label in country_df['Label'].unique():

        # Countries
        d1 = country_df[country_df['Label'] == label]

        fig.add_bar(
            x=d1['Score'],
            y=d1['production_countries'],
            orientation='h',
            name=label,
            marker=dict(
                color=colors[label],
                line=dict(color='white', width=0.5)
            ),
            width=0.22,   # thicker bars
            legendgroup=label,
            showlegend=True,
            row=1, col=1
        )

        # Genres
        d2 = genre_df[genre_df['Label'] == label]

        fig.add_bar(
            x=d2['Score'],
            y=d2['genres'],
            orientation='h',
            name=label,
            marker=dict(
                color=colors[label],
                line=dict(color='white', width=0.5)
            ),
            width=0.22,
            legendgroup=label,
            showlegend=False,
            row=1, col=2
        )

    # ------------------------------------------------
    # Axis Order
    # ------------------------------------------------
    fig.update_yaxes(
        categoryorder='array',
        categoryarray=country_order,
        row=1, col=1
    )

    fig.update_yaxes(
        categoryorder='array',
        categoryarray=genre_order,
        row=1, col=2
    )

    # ------------------------------------------------
    # Layout
    # ------------------------------------------------
    fig.update_layout(
        template='plotly_dark',
        plot_bgcolor='#0F171E',
        paper_bgcolor='#0F171E',
        font=dict(color='white'),

        title=f"{title_name} Analysis Bottom (Mean vs Q3 vs Upper Bound)",
        title_x=0.5,

        barmode='group',

        # spacing
        bargap=0.55,
        bargroupgap=0.3,

        height=750,
        width=1100,

        legend=dict(
            orientation='h',
            y=-0.18,
            x=0.5,
            xanchor='center'
        )
    )

    fig.show()
````

````python
def plot_top10_by_metric(df, metric):

    # -------------------------------
    # Prepare Data
    # -------------------------------
    data = df[['name', 'type', metric]].dropna().copy()
    data['type'] = data['type'].str.strip().str.upper()

    # -------------------------------
    # Top 10 Movies & Shows
    # -------------------------------
    top_movies = data[data['type'] == 'MOVIE'].nlargest(10, metric)
    top_shows  = data[data['type'] == 'SHOW'].nlargest(10, metric)

    top_combined = pd.concat([top_movies, top_shows])

    # -------------------------------
    # Sort for correct ranking
    # -------------------------------
    top_combined = top_combined.sort_values(by=metric, ascending=True)

    # -------------------------------
    # New Custom Colors
    # -------------------------------
    color_map = {
        'MOVIE': '#2EC4B6',   # Teal
        'SHOW': '#9D4EDD'     # Purple
    }

    # -------------------------------
    # Plot
    # -------------------------------
    fig = px.bar(
        top_combined,
        x=metric,
        y='name',
        color='type',
        color_discrete_map=color_map,
        orientation='h',
        title=f"Top 10 Movies & Shows by {metric}"
    )

    # Keep order
    fig.update_yaxes(
        categoryorder='array',
        categoryarray=top_combined['name']
    )

    # Styling
    fig.update_layout(
        template='plotly_dark',
        paper_bgcolor='#0F171E',
        plot_bgcolor='#0F171E',
        font=dict(color='white'),
        title_x=0.5,
        height=600
    )

    fig.show()
````
